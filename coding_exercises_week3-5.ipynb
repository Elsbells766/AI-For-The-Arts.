### What are Python libraries?

Python libraries are a collection of pre written code that anyone can use. Aditionally they are useful as developers don't have to write all this code from scratch.

Some commmon libraries include pandas, numpy (for maths and data analysis)scikit-learn, tensorflow, pytorch (Machine Learning), flask, django (web Development) matplotlib and seaborn(Visualization) 



```python

from dataclasses import dataclass

@dataclass
class Present:
    length: float  # in cm
    width: float   # in cm
    height: float  # in cm
    name: str = "Present"


@dataclass
class WrappingPaper:
    length: float  # in cm
    width: float   # in cm
    is_cut: bool = False


@dataclass
class Tape:
    strips_available: int  # simple counter of tape pieces you can use


@dataclass
class Scissors:
    sharp: bool = True


@dataclass
class WrappedPresent:
    present_name: str
    neatness_score: int  # a playful metric


def has_enough_paper(present: Present, paper: WrappingPaper, margin: float = 5.0) -> bool:
    """
    Check if the paper is large enough to cover the present with a margin for folding.
    For a basic rectangular box: you typically need:
      - Paper width >= (present width + 2 * present height + margin)
      - Paper length >= (present length + 2 * present height + margin)
    """
    required_width = present.width + 2 * present.height + margin
    required_length = present.length + 2 * present.height + margin
    return paper.width >= required_width and paper.length >= required_length


def measure_wrapping_paper(present: Present, margin: float = 5.0) -> dict:
    """
    Compute a recommended cut size for the wrapping paper.
    """
    recommended_width = present.width + 2 * present.height + margin
    recommended_length = present.length + 2 * present.height + margin
    return {
        "recommended_width": recommended_width,
        "recommended_length": recommended_length
    }


def cut_wrapping_paper(paper: WrappingPaper, scissors: Scissors, target_length: float, target_width: float) -> WrappingPaper:
    """
    Cut the wrapping paper to the target size using scissors.
    """
    if not scissors.sharp:
        raise RuntimeError("Scissors are not sharp enough to cut the paper neatly.")
    if target_length > paper.length or target_width > paper.width:
        raise ValueError("Target cut size is larger than the current paper.")

    # Simulate cutting: we just set the new dimensions and flag
    paper.length = target_length
    paper.width = target_width
    paper.is_cut = True
    print(f"Cut paper to {paper.length:.1f}cm x {paper.width:.1f}cm.")
    return paper


def place_present_on_paper_center(present: Present, paper: WrappingPaper) -> None:
    """
    Place the present at the center of the cut paper (conceptual step).
    """
    if not paper.is_cut:
        print("Warning: Paper was not cut to recommended size. Proceeding anyway.")
    print(f"Placed '{present.name}' at the center of the paper.")


def fold_long_sides_and_tape(paper: WrappingPaper, tape: Tape) -> None:
    """
    Fold both long sides over the present and tape to secure.
    """
    if tape.strips_available < 1:
        raise RuntimeError("No tape left to secure the long sides.")
    print("Folded one long side over the present (holding)...")
    print("Folded the other long side over the present (holding)...")
    tape.strips_available -= 1
    print("Applied tape to secure the long sides.")


def fold_short_end_and_tape(end_label: str, tape: Tape) -> None:
    """
    Fold one short end and tape it.
    """
    if tape.strips_available < 1:
        raise RuntimeError(f"No tape left to secure the {end_label} short end.")
    print(f"Folded the {end_label} short end over the taped long sides (holding)...")
    tape.strips_available -= 1
    print(f"Applied tape to secure the {end_label} short end.")


def wrap_present(present: Present, paper: WrappingPaper, scissors: Scissors, tape: Tape) -> WrappedPresent:
    """
    Orchestrates the full wrapping process and returns a WrappedPresent object.
    """
    print("=== WRAP START ===")
    print(f"Inputs: {present}, {paper}, tape strips={tape.strips_available}, scissors sharp={scissors.sharp}")

    # Measure
    plan = measure_wrapping_paper(present)
    print(f"Measured: recommended cut {plan['recommended_length']:.1f}cm x {plan['recommended_width']:.1f}cm (L x W).")

    # Check current paper size and cut accordingly
    # If current paper is larger than recommended, cut down. If smaller, error.
    if not has_enough_paper(present, paper, margin=5.0):
        raise ValueError("Wrapping paper is too small for this present. Get a larger sheet.")

    target_length = min(paper.length, plan["recommended_length"])
    target_width = min(paper.width, plan["recommended_width"])
    paper = cut_wrapping_paper(paper, scissors, target_length, target_width)

    # Place present
    place_present_on_paper_center(present, paper)

    # Fold long sides and tape
    fold_long_sides_and_tape(paper, tape)

    # Fold short ends (the “open ends”) and tape each
    fold_short_end_and_tape("first", tape)
    fold_short_end_and_tape("second", tape)

    # Simple neatness metric: more tape strips left implies efficient use (just playful)
    neatness = max(1, min(10, 10 - (3 - tape.strips_available)))  # started by using 3 strips

    print("Output: Wrapped present.")
```
